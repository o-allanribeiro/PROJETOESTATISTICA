```python
# Importando a biblioteca pandas e matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Criando um exemplo de dados com atributos categorizados e quantitativos
dados = {
    "Nome": ["Ana", "Bruno", "Carlos", "Daniela"],
    "Sexo": ["F", "M", "M", "F"],
    "Cor dos olhos": ["Azul", "Castanho", "Verde", "Castanho"],
    "Estado civil": ["Solteira", "Casado", "Divorciado", "Viúva"],
    "Altura": [1.65, 1.75, 1.80, 1.60],
    "Peso": [60, 80, 85, 55],
    "Idade": [25, 32, 40, 50],
    "Renda": [3000, 5000, 4000, 2000]
}
```

```python
df = pd.DataFrame(dados)
print(df)
```

```python
fig, ax = plt.subplots(figsize=(16, 2)) # adjust the size of the figure
ax.axis('off')

pd.plotting.table(ax, df, loc='center')  # where df is your data frame
fig.suptitle('Dados categoricos e dados quantitativos') # add a title to the figure
plt.savefig('mytable.png', bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(10, 1.4)) # adjust the size of the figure
ax.axis('off')

table = pd.plotting.table(ax, df, loc='center')  # where df is your data frame

# change the font and alignment of the cells
for cell in table.get_celld().values():
    cell.set_fontsize(12)
    cell.set_text_props(fontname='Times New Roman', ha='center')

fig.suptitle('Dados categóricos e quantitativos', fontsize=14) # add a title to the figure

plt.savefig('Imagens/mytabletypedata.png', bbox_inches='tight')
```
