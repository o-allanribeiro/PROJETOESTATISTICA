# Importando a biblioteca pandas e matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Um exemplo de dados com atributos qualitativos e quantitativos

```python

dados = {
    "Nome": ["Ana", "Bruno", "Carlos", "Daniela"],
    "Sexo": ["F", "M", "M", "F"],
    "Cor dos olhos": ["Azul", "Castanho", "Verde", "Castanho"],
    "Estado civil": ["Solteira", "Casado", "Divorciado", "Viúva"],
    "Altura": [1.65, 1.75, 1.80, 1.60],
    "Peso": [60, 80, 85, 55],
    "Idade": [25, 32, 40, 50],
    "Renda": [3000, 5000, 4000, 2000]
}
```

## Impressão dos dados

```python
df = pd.DataFrame(dados)
print(df)
```

### Ajuste do tamanho da figura, Definir o dataframe, mudar as fontes, alinhar as celulas e adicionar um título

```python
#Ajuste das figuras
fig, ax = plt.subplots(figsize=(10, 1.4))
ax.axis('off')

#Definição do dataframe para a tabela com o matplotlib
table = pd.plotting.table(ax, df, loc='center')

#Personalizar as fontes e celulas
for cell in table.get_celld().values():
    cell.set_fontsize(12)
    cell.set_text_props(fontname='Times New Roman', ha='center')

#Adicionar um título para a tabela
fig.suptitle('Dados categóricos e quantitativos', fontsize=14)

plt.savefig('Imagens/mytabletypedata1.png')
```
