# Importando bibliotecas

```python
import matplotlib.pyplot as plt
```

```python
# dados para plotar
labels = ['Python', 'C++', 'Java', 'Dart', 'C#', 'JavaScript']
sizes = [51, 62, 63, 54, 65, 65]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','pink','orange']
explode = (0.1, 0.1 ,0 ,0 ,0 ,0) # explode a fatia do Python e do C++


```

```python

# plot gráfico de rosca
plt.pie(sizes,
        explode=explode,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',
        shadow=True,
        startangle=140)

# desenhar um círculo branco no centro
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title ( 'Pesquisa de preferência - Evento 1' )

# ajustar o eixo para ficar igual
plt.axis('equal')

# mostrar o gráfico
plt.show()
```
