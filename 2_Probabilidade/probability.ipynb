# Probabilidades

### Importando bibliotecas

```python
# Importar a biblioteca random
import random
# Importar a biblioteca matplotlib
import matplotlib.pyplot as plt
# Importar a biblioteca numpy
import numpy as np
# Importar a biblioteca pandas
import pandas as pd
```

## Definindo funções

### Definir uma função para simular o lançamento de uma moeda

```python
# Definindo a função para simular o lançamento de uma moeda
def lancar_moeda():
  # Gerar um número aleatório entre 0 e 1
  x = random.random()
  # Se o número for menor que 0.5, retornar "cara"
  if x < 0.5:
    return "cara"
  # Senão, retornar "coroa"
  else:
    return "coroa"
```

### Definir uma função para simular o lançamento de um dado

```python
# Definindo a função para simular o lançamento de um dado
def lancar_dado():
  # Gerar um número inteiro aleatório entre 1 e 6
  y = random.randint(1,6)
  # Retornar o número gerado
  return y
```

### Definir uma função para simular o experimento de lançar uma moeda e depois lançar um dado

```python
# Definindo a função para simular o experimento
def experimento():
  # Chamar a função lancar_moeda e armazenar o resultado em uma variável
  moeda = lancar_moeda()
  # Chamar a função lancar_dado e armazenar o resultado em outra variável
  dado = lancar_dado()
  # Retornar uma tupla com os resultados da moeda e do dado
  return (moeda, dado)
```

### Simulação de experimentos aleatórios com Python

```python
# Definindo a função para simular o experimento N vezes
def experimento_N(N):
    # Criar uma lista vazia para armazenar os resultados
    resultados = []
    # Repetir o experimento N vezes
    for i in range(N):
        # Chamar a função experimento e armazenar o resultado em uma variável
        resultado = experimento()
        # Adicionar o resultado à lista de resultados
        resultados.append(resultado)
    # Retornar a lista de resultados
    return resultados
```

### Simulação de experimentos aleatórios com Python

```python
# Definindo a função para calcular a frequência de ocorrência de cada resultado
def frequencia(resultados):
    # Criar um dicionário vazio para armazenar as frequências
    freq = {}
    # Para cada resultado na lista de resultados
    for resultado in resultados:
        # Se o resultado já estiver no dicionário
        if resultado in freq:
            # Incrementar a frequência
            freq[resultado] += 1
        # Senão
        else:
            # Inicializar a frequência com 1
            freq[resultado] = 1
    # Retornar o dicionário de frequências
    return freq
```

# Exercícios de fixação - Probabilidades

```python
# Definindo a função para calcular a probabilidade de ocorrência de cada resultado
def probabilidade(freq, N):
    # Criar um dicionário vazio para armazenar as probabilidades
    prob = {}
    # Para cada resultado na lista de resultados
    for resultado in freq:
        # Calcular a probabilidade de ocorrência do resultado
        prob[resultado] = freq[resultado] / N
    # Retornar o dicionário de probabilidades
    return prob
```

### Exercícios de fixação - Probabilidades

1. Simule o lançamento de uma moeda 100 vezes e calcule a frequência de ocorrência de cada resultado.
    - Em seguida, calcule a probabilidade de ocorrência de cada resultado.

```python
# Simular o lançamento de uma moeda 100 vezes
resultados = experimento_N(100)

# Calcular a frequência de ocorrência de cada resultado
freq = frequencia(resultados)

# Calcular a probabilidade de ocorrência de cada resultado
prob = probabilidade(freq, 100)

# Imprimir a frequência de ocorrência de cada resultado
print("Frequência de ocorrência de cada resultado:")
print(freq)
```

```python
# Imprimir a probabilidade de ocorrência de cada resultado
print("Probabilidade de ocorrência de cada resultado:")
print(prob)
```

```python
# Visualizar em um gráfico de barras a frequência de ocorrência de cada resultado

# Criar uma lista com os resultados
resultados = list(freq.keys())

# Criar uma lista com as frequências
frequencias = list(freq.values())

# Criar uma lista com as probabilidades
probabilidades = list(prob.values())

# Criar uma lista com os valores de x
x = np.arange(len(resultados))

# Criar uma figura
fig, ax = plt.subplots()

# Criar as barras
rects1 = ax.bar(x - 0.25, frequencias, 0.25, label='Frequência')
rects2 = ax.bar(x, probabilidades, 0.25, label='Probabilidade')

# Tamanho da fonte dos rótulos
plt.rcParams.update({'font.size': 14})

# tamanho da fonte dos números no eixo x
plt.rcParams['xtick.labelsize'] = 14

# tamanho do gráfico em pixels
fig.set_size_inches(20, 12)

# Adicionar rótulos de texto nas barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adicionar rótulos de texto nas barras
autolabel(rects1)
autolabel(rects2)

# Adicionar título ao gráfico
ax.set_title('Frequência e probabilidade de ocorrência de cada resultado')

# Adicionar rótulos no eixo x
ax.set_xticks(x)
ax.set_xticklabels(resultados)

# Adicionar legenda
ax.legend()

# Exibir o gráfico
plt.show()

# Salvar o gráfico em um arquivo PNG
fig.savefig('frequencia_probabilidade.png')
```
