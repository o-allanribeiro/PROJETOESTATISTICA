### Importando as bibliotecas necessárias

```python
import matplotlib.pyplot as plt 
import numpy as np 
import random
```

### Definindo a função que gera os círculos
- n é o número de círculos
- raio é o raio dos círculos
- cor é a cor dos círculos

#### Retorna uma lista de objetos Circle

#### Gerando coordenadas aleatórias para o centro do círculo
- Criando o objeto Circle com as coordenadas, o raio e a cor
- Adicionando o círculo à lista 

```python
def gerar_circulos(n, raio, cor):
    circulos = []
    for i in range(n):
        # Gerando coordenadas aleatórias para o centro do círculo
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        # Criando o objeto Circle com as coordenadas, o raio e a cor
        circulo = plt.Circle((x, y), raio, color=cor)
        # Adicionando o círculo à lista
        circulos.append(circulo)
    # Retornando a lista de círculos
    return circulos
```

### Definindo a população como um conjunto de 100 círculos azuis de raio 0.05

```python
populacao = gerar_circulos(100, 0.05, 'blue')
```


### Definindo a amostra como um conjunto de 20 círculos vermelhos de raio 0.05


```python
amostra = gerar_circulos(20, 0.05, 'red')
```

```python
#Criando a figura e os eixos
fig, ax = plt.subplots()
plt.title ( 'Demonstração de dados - População e amostra' )
#Definindo os limites dos eixos
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)



#Desenhando os círculos da população e da amostra na figura
for circulo in populacao:
    ax.add_artist(circulo)
    for circulo in amostra: 
        ax.add_artist(circulo)

#Mostrando a figura na tela

plt.savefig(r"C:\Users\allan\PROJETOESTATISTICA\Imagens\readme\samp_popu")
plt.show()
```
