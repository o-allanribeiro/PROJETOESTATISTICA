# Importando bibliotecas

```python
import matplotlib.pyplot as plt
import random

```

### Criar uma lista com 50 números aleatórios entre 15 e 65

```python
data = [random.randint(15, 65) for i in range(50)]
# Mostrar a lista
data
```

```python
# criando intervalos de classe
classInterval = [15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65]

# plotando o histograma
plt.hist(data, bins=classInterval)

# formatando o gráfico
plt.title('Pesquisa idade aleatória')
plt.xlabel('Itar')
plt.ylabel('Frequência')

# mostrando o gráfico
plt.show()
```
